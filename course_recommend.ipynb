```python
import re
def clean_text(text):
  text = str(text).lower()
  text = re.sub('\[.*?\]', '', text)
  text = text.strip()
  return text

df['Course Name'] = df['Course Name'].apply(clean_text)
df.head()
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
nltk.download('wordnet')

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    text = clean_text(text)
    words = text.split()
    words = [lemmatizer.lemmatize(w) for w in words if w not in stop_words]
    return " ".join(words)

df['Course Name'] = df['Course Name'].apply(preprocess_text)
```

```python
df.head()
```

```python
from sentence_transformers import SentenceTransformer, util
import torch

model = SentenceTransformer('all-MiniLM-L6-v2')
course_embeddings = model.encode(df['Course Name'].tolist(), convert_to_tensor=True)
```

```python
def recommend(query, top_k=5):
    query_clean = preprocess_text(query)
    query_embedding = model.encode(query_clean, convert_to_tensor=True)
    similarities = util.cos_sim(query_embedding, course_embeddings)[0]

    top_results = torch.topk(similarities, k=top_k)

    print(f"🔍 คำค้น: {query}\n📚 คอร์สที่แนะนำ:")
    for score, idx in zip(top_results.values, top_results.indices):
        course = df.iloc[idx.item()]
        print(f"\n⭐ {course['Course Name']}")
        print(f"🎯 ระดับ: {course['Difficulty Level']}, 🌟 เรตติ้ง: {course['Course Rating']}")
        print(f"🔗 {course['Course URL']}")
```

```python
recommend("math")
```

```python
import torch
import os

# สร้างโฟลเดอร์สำหรับบันทึก
export_dir = "export_course_model"
os.makedirs(export_dir, exist_ok=True)

# 1. บันทึกโมเดล SentenceTransformer
model.save(f"{export_dir}/sentence_model")

# 2. บันทึกเวกเตอร์ embeddings
torch.save(course_embeddings, f"{export_dir}/course_embeddings.pt")

# 3. บันทึก DataFrame ที่มี cleaned_name
df.to_csv(f"{export_dir}/course_data.csv", index=False)

print("✅ โมเดลและข้อมูลถูกบันทึกเรียบร้อยแล้วในโฟลเดอร์:", export_dir)
```
